```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data = {
    'Year': list(range(2010, 2025)), 
    'Rainfall': [850, 900, 870, 880, 890, 910, 920, 940, 930, 950, 970, 980, 990, 1000, 1020]
}

df = pd.DataFrame(data)

X = df[['Year']]  
y = df['Rainfall']

# Create and train the Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Predict rainfall for 2026
predicted_rainfall = model.predict([[2026]])
print(f"Predicted Rainfall for 2026: {predicted_rainfall[0]:.2f} mm")

# Plotting the data and regression line
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.scatter(2026, predicted_rainfall, color='green', label='Prediction for 2026', zorder=5)
plt.xlabel("Year")
plt.ylabel("Rainfall (mm)")
plt.title("Year vs Rainfall - Linear Regression")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
