```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data: Year vs Rainfall (15 years only)
X = [[2010], [2011], [2012], [2013], [2014], [2015], [2016], [2017], [2018], [2019],
     [2020], [2021], [2022], [2023], [2024]]
Y = [2.5, 2.52, 2.54, 2.48, 2.52, 2.54, 2.55, 2.7, 2.9, 3.2, 3.16, 3.28, 3.2, 3.15, 3.26]

# Create and train the model
model = LinearRegression()
model.fit(X, Y)

# Predict for 2026
pred_2026 = model.predict([[2026]])
print(f"Predicted rainfall for the year 2026: {pred_2026[0]:.2f} mm")

# Plotting
plt.scatter(X, Y, color='blue', label='Actual data')
plt.plot(X, model.predict(X), color='red', label='Regression line')
plt.scatter([2026], pred_2026, color='green', label='Prediction (2026)')
plt.xlabel('Year')
plt.ylabel('Rainfall (mm)')
plt.title('Year vs Rainfall')
plt.legend()
plt.grid(True)
plt.show()
```
